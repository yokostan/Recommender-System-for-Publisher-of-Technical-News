```python
#import word-press dataset
import pandas as pd
import numpy as np

file_name = 'word2vec_vector_norm.csv'

df = pd.read_csv('./' + file_name)
df
```

```python
from sklearn.cluster import KMeans
import numpy as np
kmeans = KMeans(n_clusters=30, random_state=0).fit(df)
```

```python
cluster_map = pd.DataFrame()
cluster_map['data_index'] = df.index.values
cluster_map['cluster'] = kmeans.labels_
```

```python
cluster_map
```

```python
from matplotlib import pyplot as plt
plt.scatter(cluster_map.data_index, 175000*np.random.random_sample((176359, )), c=cluster_map.cluster)
plt.show()
```

```python
max(cluster_map.cluster)
```

```python
file_name = 'wordpress_word2vec_splitted.csv'

df2 = pd.read_csv('./' + file_name)
df2
```

```python
df3 = df2.assign(cluster_id = cluster_map.cluster)
df3
```

```python
#saving our normalized splitted word2vec
df3.to_csv("./word2vec_cluster_kmeans30_total.csv")
```
