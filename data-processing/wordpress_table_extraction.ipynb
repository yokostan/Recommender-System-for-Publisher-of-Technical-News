```python
from google.cloud import storage
from io import BytesIO
import pandas as pd
import numpy as np

client = storage.Client()
bucket = 'digital-trend-data-unzipped'
medium_path = 'wordpress'
df_res = []

for i in range(32):
    if i < 10:
        idx = '0' + str(i)
    else:
        idx = str(i)

    file_name =  '00' + idx + '_part_00' +'.csv'

    !gsutil -m cp gs://{bucket}/{medium_path}/{file_name} ./{file_name}
    df_raw = pd.read_csv('./' + file_name)

    # select columns
    # post_id/ date / category / pri_category / second_category / tag / title / url
    df_selected = df_raw.iloc[:, [0, 1, 4, 6, 7, 9, 10, 22]]
    df_selected = df_selected.astype(object).replace(np.nan, '')
    df_selected.iloc[:, 2] = df_selected.iloc[:, 2] +'|'+ df_selected.iloc[:, 3] +'|'+ df_selected.iloc[:, 4] +'|'+ df_selected.iloc[:, 5]
    
    df_res.append(df_selected.iloc[:, [0, 1, 2, 6, 7]])
    
result = pd.concat(df_res)
print(result.shape)
```

```python
file = open('./wordpress_combine.csv',"w")
result.to_csv(file, index=False)
!gsutil cp ./wordpress_combine.csv gs://{bucket}/{medium_path}/
```
