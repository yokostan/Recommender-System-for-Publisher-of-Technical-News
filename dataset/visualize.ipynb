```
import altair as alt
import pandas as pd
import numpy as np
from google.colab import files
uploaded = files.upload()
```

```
df = pd.read_excel('./cluster_overview.xlsx')
df
```

```
df = pd.DataFrame(df)
```

```
alt.Chart(df).mark_bar().encode(
    x = 'cluster_id',
    y = 'number'
)
```

```
df = df.assign(y = np.random.randint(0, 20, 50))
df.head()
```

```
alt.Chart(df).mark_circle().encode(
    alt.X('cluster_id', axis=None),
    alt.Y('y', axis=None),
    size = 'number',
    color=alt.Color('cluster_id', legend=alt.Legend(title="Species by color"))
).properties(
    width=600,
    height=600
).mark_text(
    align='left',
    baseline='middle',
    dx=7
).encode(
    text='topic'
)
```

```
from altair import *
range_ = ['red', 'green', 'blue', 'purple','darkblue']

alt.Chart(df).mark_text(align='left', baseline='middle', dy=4).encode(
    alt.X('y', axis=None, scale=Scale(domain=[-10, 150])),
    alt.Y('cluster_id', axis=None, scale=Scale(domain=[-0.5, 50.5])),
    text = 'topic',
    size = 'number',
    color= alt.Color('cluster_id', scale=alt.Scale(range=range_))
).properties(
    width=800,
    height=1000,
    title='50 content clusters from Digital Trends 170000+ posts'
).configure_title(
    fontSize=25,
    color='black',
    align='center',
    fontStyle='Times'
)
```
